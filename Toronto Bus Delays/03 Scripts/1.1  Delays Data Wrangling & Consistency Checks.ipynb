### 02 Addressing mixed data types

```python
#function to find any columns with mixed data types
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
#changing route data type to string
df['Route'] = df['Route'].astype('str')
```

```python
#changing direction data type to string
df['Direction'] = df['Direction'].astype('str')
```

```python
#checking again for mixed data types
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

##### All mixed columns have been dealt with

### 03 Addressing duplicates

```python
#creating set containing only duplicate values
df_dup=df[df.duplicated()]
df_dup
```

```python
#there are 154 duplicate entries, exporting them to csv
df_dup.to_csv(os.path.join(path, '02 Data','Prepared', 'duplicates.csv'))
```

```python
#creating copy of dataset without duplicates
df2=df.drop_duplicates()
```

### 04 Addressing missing data 

```python
df2
```

```python
#looking for missing values
df2.isnull().sum()
#missing values in direction table are saved as nan
```

```python
#locating any nan values
dfnan=df2[df2.eq("nan").any(1)]
dfnan
```

```python
#dropping direction column
df2.drop(['Direction'], axis=1, inplace=True)
```

```python
#locating any other nan values
dfnan=df2[df2.eq("nan").any(1)]
dfnan
```

#### remaining missing values are in the Route column

```python
#exporting dfnan to csv
dfnan.to_csv(os.path.join(path, '02 Data','Prepared', 'nan_entries.csv'))
```

```python
#deleting all nan rows
df3 = df2.drop(dfnan.index)
```

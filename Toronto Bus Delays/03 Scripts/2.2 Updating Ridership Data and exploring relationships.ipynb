## 1 Importing libraries/datasets

```python
#import librareis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
#importing datasets
path=r'/Users/Amaikuru/Desktop/Toronto Bus Delays'
df=pd.read_pickle(os.path.join(path, '02 Data', 'Prepared', 'delays_final.pkl'))
```

```python
#importing ridership dataset
ride=pd.read_csv(os.path.join(path, '02 Data', 'Original', 'monthly ridership ttc.csv'))
```

## 2 Updating ridership column

```python
pd.set_option('display.max_columns', None)
```

```python
df.columns
```

```python
df.head()
```

```python
ride.head()
```

```python
#dropping ridership column
df.drop(columns='Ridership',axis=1,inplace=True)
```

```python
df.shape
```

```python
#joining with year and month columns
df2=df.merge(ride, left_on=['month','year'],right_on=['Month','Year'],how='inner')
```

```python
df2.head()
```

```python
#dropping extra columns
df2.drop(columns=['Year','Month'],axis=1,inplace=True)
```

```python
df2.shape
```

## 3 Exploring relationships

```python
sub=df2[['Ridership','month_delay_count','avg_temperature']]
```

```python
#displaying heat map
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(30,30))
# Creating the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.title('Correlation matrix', fontsize=16)
```

## 1 Importing libraries/ dataset

## 2 Wrangling for analysis

```python
df.columns
```

```python
df.head()
```

```python
df.dtypes
```

```python
#selecting only relevant numerical variables
df2=df[['Min Delay', 'Min Gap', 'ward_population', 'ward_median_age',
       'ward_median_income', 'ward_density','child_pop','youth_pop','adult_pop','senior_pop','ward_low_income','n_poi', 'n_schools', 'n_worship',
       'dist_to_nearest_station', 'n_stations', 'n_stops', 'n_stops_route',
       'stop_delay_count', 'route_delay_count', 'ward_delay_count']]
```

```python
df2.head()
```

```python
pd.set_option('display.max_columns',None)
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
#null values are outside wards
#replacing outside wards values with 0
df2.fillna(0,inplace=True)
```

```python
df2.isnull().sum()
```

```python
#changing mm to m in df2
df2['dist_to_nearest_station']=df2['dist_to_nearest_station']/1000
```

```python
#scaling data to reduce variance
scaler=StandardScaler().fit(df2)
```

```python
df_std=scaler.fit_transform(df2)
```

## 3 Elbow technique

```python
# Import ElbowVisualizer
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
# k is range of number of clusters.
visualizer = KElbowVisualizer(model, k=(2,30), timings= True)
visualizer.fit(df2)        # Fit data to visualizer
visualizer.show()   
```

#### the elbow method shows the optimal number of clusters is 7

## 4 K-means clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 7)
```

```python
# Fit the k-means object to the data.

kmeans.fit(df_std)
```

```python
#incuding resulting clusters in dataset
df2['clusters'] = kmeans.fit_predict(df_std)
```

```python
df2.head()
```

```python
df2['clusters'].value_counts()
```

#### most rows belong to cluster 2 currently, the least rows belong to cluster 5

## 5 Visualizing clusters

```python
#visualizing distance to nearest station and number of delays on stop


plt.figure(figsize=(20,10))
ax = sns.scatterplot(x=df2['dist_to_nearest_station'], y=df2['stop_delay_count'], hue=kmeans.labels_,palette="tab10" ,s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('dist_to_nearest_station') # Label x-axis.
plt.ylabel('stop_delay_count') # Label y-axis.
plt.show()
```

#### cluster 2 has the stops with the most delays, some of these stops are located at stations, the farthest stops are located less than 50 m from the nearest station, majority of the stops are closer to stations


#### cluster 0 also has stops with a high number of delays and stops located at stations but cluster 0 only goes as far at 30 m from the nearest station (stops are mostly closer to stations)

#### cluster 3 has stops with a wide range of distances, from 0 to about 78, the highest number of delays in this cluster is 250, a majority of these stops are located far away from stations

#### cluster 1 also has stops with a wide range of distances ranging from 0 to more than 80m, this cluster has the farthest stops and a majority of the stops are located relatively far away from stations. This cluster has delay counts that are relatively higher than those with similar distances

#### cluster 4 has stops located mainly between  20 and 60 m from the nearest station, these stops have fewer delays than others

#### cluster 5 has very few stops and they are all less than 40m to the nearest station and have the very few delays

#### cluster 6 also has stops located closer to stations and have very few delays

#### these make sense because, stations would naturally have the most passengers and so stops at stations would have more delays while stops far from stations will have fewer delays
